# Create and Access SQLite Database Using Python

## Objective
In this lab, I will:
- Create a SQLite database using Python
- Create a table called INSTRUCTOR
- Insert data into the table
- Query and update the data
- Load the data into pandas

## Step 1: Import Required Library

We begin by importing the sqlite3 library, which allows Python to interact with SQLite databases.

```python
import sqlite3
```

## Step 2: Create a Database Connection

Here we create a connection to a database called INSTRUCTOR.db.

If the database does not exist, SQLite will automatically create it.

```python
conn = sqlite3.connect("INSTRUCTOR.db")
```

## Step 3: Create a Cursor Object

The cursor is used to execute SQL commands such as creating tables and inserting data.

```python
cursor_obj = conn.cursor()
```

## Step 4: Create the INSTRUCTOR Table

We now create a table called INSTRUCTOR with the following columns:
- ID (Primary Key)
- First Name
- Last Name
- City
- Country Code

```python
cursor_obj.execute("""
CREATE TABLE IF NOT EXISTS INSTRUCTOR(
ID INTEGER PRIMARY KEY NOT NULL,
FNAME VARCHAR(20),
LNAME VARCHAR(20),
CITY VARCHAR(20),
CCODE CHAR(2)
)
""")

print("Table created successfully")
```

## Step 5: Insert Data into the Table

We will insert sample records into the INSTRUCTOR table.

```python
cursor_obj.execute("INSERT INTO INSTRUCTOR VALUES (1, 'Rav', 'Ahuja', 'TORONTO', 'CA')")
cursor_obj.execute("INSERT INTO INSTRUCTOR VALUES (2, 'Raul', 'Chong', 'Markham', 'CA')")
cursor_obj.execute("INSERT INTO INSTRUCTOR VALUES (3, 'Hima', 'Vasudevan', 'Chicago', 'US')")

conn.commit()

print("Data inserted successfully")
```

We use commit() to save the changes permanently to the database.
## Step 6: Query Data from the Table

We use a SELECT statement to retrieve all records from the INSTRUCTOR table.

```python
cursor_obj.execute("SELECT * FROM INSTRUCTOR")

rows = cursor_obj.fetchall()

print("All data:")
for row in rows:
    print(row)
```

## Step 7: Fetch One Record

We can also retrieve a limited number of rows using fetchmany().

```python
cursor_obj.execute("SELECT * FROM INSTRUCTOR")

one_row = cursor_obj.fetchmany(1)

print("First row:")
print(one_row)
```

## Step 8: Update Data in the Table

We update the city of Rav from TORONTO to MOOSETOWN.

```python
cursor_obj.execute(
    "UPDATE INSTRUCTOR SET CITY='MOOSETOWN' WHERE FNAME='Rav'"
)

conn.commit()
```

## Verify Update

```python
cursor_obj.execute("SELECT * FROM INSTRUCTOR")

rows = cursor_obj.fetchall()

for row in rows:
    print(row)
```

## Step 9: Load Data into Pandas

We use pandas to convert the SQL table into a DataFrame for analysis.

```python
import pandas as pd

df = pd.read_sql_query("SELECT * FROM INSTRUCTOR", conn)

df
```
